```python
import pandas as pd
```

```python
df = pd.read_csv("nfl_player_av_state.csv").drop(columns = ['percentile_av', 'Player', 'Unnamed: 0', 'Tm'])
df.head()
```

```python
for i in range(len(df)):
    df.iat[i, 0] = df.iat[i, 0].split('-')[0]
```

```python
df = df.groupby(['Year', 'state']).mean()
```

```python
df.reset_index(inplace = True)
```

```python
df.to_csv('year_state_av_avg.csv', index = False)
```
