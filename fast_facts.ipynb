```python
import pandas as pd
```

```python
mlb_stats = pd.read_csv("libs/unclean_data/mlb_war_ratings/mlb_war_ratings_all.csv")
mlb_stats.head(5)
```

```python
for i in range(len(mlb_stats)):
    mlb_stats.iat[i, 2] = mlb_stats.iat[i, 2].split('\\')[0]
```

```python
mlb_stats
```

```python
mlb_salary = pd.read_csv("libs/unclean_data/mlb_player_salaries/mlb_player_salaries_all.csv")
mlb_salary.head(5)
```

```python
for i in range(len(mlb_salary)):
    first, x, y = mlb_salary.iat[i, 3][0].strip().split('\n')
#     print(first)
#     print(last)
    mlb_salary.iat[i, 3] = first
```

```python
mlb_salary.head(5)
```

```python
mlb_salary = mlb_salary.rename(columns={"name":"Player"})
```

```python
mlb_merge = pd.merge(mlb_stats, mlb_salary, on='Player')
```

```python
mlb_merge.head(5)
```

```python
mlb_merge.to_csv("libs/unclean_data/mlb_fast_facts_data.csv")
```
