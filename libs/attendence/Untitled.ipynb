```python
import pandas as pd
```

```python
mlb_data = pd.read_csv("mlb_attendance.csv")
nfl_data = pd.read_csv("nfl_attendance.csv")
nba_data = pd.read_csv("nba_attendance.csv")
```

```python
mlb_data=mlb_data[['0','1','2','3','4','5','year']]
mlb_data["sport"]="mlb"
mlb_data
nfl_data=nfl_data[['0','1','2','3','4','5','year']]
nfl_data["sport"]="nfl"
nfl_data
nba_data=nba_data[['0','1','2','3','4','5','year']]
nba_data
nba_data["sport"]="nba"
```

```python
attendance = [mlb_data, nfl_data, nba_data]

all_attendance = pd.concat(attendance)
```
